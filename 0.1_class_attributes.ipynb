## Class Attributes / Variables - Examples and use cases

class attributes are variables that are same for all instances of the Class. it is defined in the beginning of the Class, and have same value for all objects.

```python
class Student:
    # Class Attribute / Variable
    school_name: str = "Government High School"

    def __init__(self, name: str, age: int):
        # Instance variables
        self.name: str = name
        self.age: int = age
```

```python
# all students have same school
s1 = Student("ahmed", 15)
s2 = Student("sara", 14)

print(s1.school_name)  # Government High School
print(s2.school_name)  # Government High School
```

### How class attributes / are used properly

```python
class Car:
    # class attribute
    wheels: int = 4

    def __init__(self, brand: str) -> None:
        self.brand: str = brand

    @classmethod
    def show_wheels(cls) -> None:
        print(f"All cars have '{cls.wheels}' wheels.")

    def __str__(self) -> str:
        return f"The car with Brand name:'{self.brand}' have {self.__class__.wheels} wheels."
```

`cls` automatically take that class method call, in which it is called.

```python
# direct call from ClassName - the wrong way
print(Car.wheels)  # print only value
print("==" * 50)

# OR through Object
my_car = Car("yaris")
print("My car ", my_car.brand, " have ", Car.wheels)  # wrong way
print("==" * 50)

# by using the default __str__ method
print(my_car)
print("==" * 50)

my_car.show_wheels()  # right way
```

```python
# Note:- this code works in Python 3.13+
class Employee:
    """
    Employee class demonstrating @classmethod as alternate constructors.
    This is the most powerful real-world use of classmethod.
    """

    company: str = "TechCorp"
    default_department: str = "General"

    def __init__(self, name: str, salary: float, department: str) -> None:
        self.name: str = name
        self.salary: float = salary
        self.department: str = department

    @classmethod
    def from_dict(cls, data: dict) -> "Employee":
        """
        Alternate constructor — create Employee from a dictionary.
        cls = Employee (or any subclass), so inheritance works perfectly.
        """
        return cls(
            name=data["name"],
            salary=data["salary"],
            department=data.get("department", cls.default_department),
        )

    @classmethod
    def from_csv_string(cls, csv_string: str) -> "Employee":
        """
        Alternate constructor — create Employee from a CSV string.
        Example: 'Ali,85000,Engineering'
        """
        parts: list[str] = csv_string.strip().split(",")
        return cls(
            name=parts[0],
            salary=float(parts[1]),
            department=parts[2] if len(parts) > 2 else cls.default_department,
        )

    @classmethod
    def get_company(cls) -> str:
        """Return company name — works correctly even in subclasses."""
        return cls.company

    def __str__(self) -> str:
        return f"{self.name} | {self.department} | PKR {self.salary:,.0f} | {self.__class__.company}"
```

```python
# ✅ Normal constructor
emp1 = Employee(name="Ali", salary=85000, department="Engineering")

# ✅ From dictionary (e.g., API response / JSON)
emp2: Employee = Employee.from_dict(
    {"name": "Sara", "salary": 90000, "department": "Design"})

# ✅ From CSV string (e.g., file reading)
emp3: Employee = Employee.from_csv_string("Hamza,75000,Marketing")

print(emp1)
print(emp2)
print(emp3)
print(f"Company: {Employee.get_company()}")
```

**🎯 Simple Rule — when to use what?**

Only required **Fixed Value**?

    ✅ YES → Direct Access: Car.wheels OR self.__class__.wheels

Need class Level  Logic/Behaviour?

    ✅ YES → @classmethod with cls

Are you using Inheritance?

    ✅ YES → always use @classmethod
