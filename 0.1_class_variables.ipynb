## Class variables - Examples and use cases

class variables are variables that are same for all instances of the Class. it is defined in the begining of the Class, and have same value for all objects.

```python
class Student:
    # Class Variable
    school_name: str = "Government High School"

    def __init__(self, name: str, age: int):
        # Instance variables
        self.name: str = name
        self.age: int = age
```

```python
# all students have same school
s1 = Student("ahmed", 15)
s2 = Student("sara", 14)

print(s1.school_name)  # Government High School
print(s2.school_name)  # Government High School
```

### How class variables can be used?

```python
# Inside class

class Car:
    wheels = 4

    def __init__(self, brand: str):
        self.brand: str = brand

    @classmethod
    def show_wheels(cls):
        print(f"All cars have {cls.wheels}.")
```

```python
# کلاس کے نام سے براہ راست
print(Car.wheels)  # 4

# یا آبجیکٹ کے ذریعے
my_car = Car("Toyota")
print("My car ", my_car.brand, " have ", Car.wheels)
```
