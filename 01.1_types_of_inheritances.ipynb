## Types of Inheritances

* `Simple` - Child -> Parent - Child Inherit from parent - Can access non-private data members and methods.
* `Heirarchical` - Child 1 & Child 2 -> Parent - Multiple Child Inherit properties from same parent. Children are not related to eachother.
* `Multilevel` - Child -> Parent -> Grandparent - Child Inherit from Parent and Parent Inherit from Grandparent. Child can access properties of both Parent and Grandparent. super() of each class can reach out to just the parent.
* `Multiple` - Child -> Parent 1 & Parent 2 - Child Inherit from multiple parents. Child can access properties of both parents. super() of child class can reach out to both parents but super() of parent class can reach out to only its parent. Not allowed in Java.
* `Hybrid` - Combination of multiple and multilevel. Not allowed in Java. It looks like Doamon Shape. Child -> Parent 1 & Parent 2 -> Both -> Grandparet.

```python
class User:
    def __init__(self):
        print("User Constructor")

        self.webiste = "www.google.com"
        self._field = "e-commerce" # protected
        self.__duration = "1 year" # private


    def login(self):
        print("User Login")

    
    def logout(self):
        print("User Logout")
```

```python
class Buyer(User):
    def __init__(self):
        print("Buyer Constructor")

        super().__init__()
        b_name = self.buyer_name = "John Doe"
        print(b_name)

    def login(self):
        print("Buyer Login")
        super().logout()
```

```python
b1 = Buyer()
print(b1)
print(b1.webiste)
print(b1._field)
# print(b1.__duration) # AttributeError: 'Buyer' object has no attribute '__duration', because __duration is private to User class and not accessible in Buyer class.
```

```python
# MUltiple Inheritance - MRO (Method Resolution Order) 
# The order in which Python looks for a method in a hierarchy of classes. 
# It follows the C3 linearization algorithm to determine the method resolution order.

class Grandparent: #Grandeparent
    def __init__(self):
        print("Grandfather init")

    def __str__(self) -> str:
        return "I am class A - Grandfather"

class Father(Grandparent): # Parent 1
    def __init__(self):
        print("Father init")
        super().__init__()

class Mother(Grandparent): # Parent 2
    def __init__(self):
        print("Mother init")
        super().__init__()

class Child(Mother, Father): # Child
    """Important Point to remember
        we have set the heirarchy, first we initiate B(father) and then C(mother). so we will got inheritance in the same same.
        for example, in real-life we will approach our favorite parent first for any help.
    """
    def __init__(self):
        print("Child init")
        super().__init__()

    def __str__(self) -> str:
        parent_str = super().__str__()
        return f"{parent_str}"

child = Child()
print(child)

Child.mro() 
#it will show the exact order in which the methods will be resolved.
# It will show the order of classes that will be searched when we call a method on an instance of class D.
```
