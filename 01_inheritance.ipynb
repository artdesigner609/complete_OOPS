```python
class User:  # parent class
    def __init__(self, username, id) -> None:
        self.username = username
        self.id = id

    def login(self):
        print(f"{self.username} logged in successfully.")

    def logout(self):
        print(f"{self.username} logged out successfully.")

    def __str__(self) -> str:
        return f"Welcome {self.username}, Your ID is:{self.id}"
```

```python
# so to inherit parent User class in the child class, we do this:
class Student(User):
    pass
```

```python
s1 = Student("John", 36)
print(s1)
s1.login()
s1.logout()
```

If we observe above, we inherit the properties of parent class. And child class runs accurately. But what if we ned to add some more properties in the child class. then we can do this by using __init__ method in the child class.

But if we do so, we will be not following the `DRY principle`. We have mentioned same `username`, `id` in the parent class, and then using same `student_name`, `id` and `age` in the child class `Student`. So how it is different? mean why we provide info in the User class and then again in the Student class?

So, to cater this, we will use `super() function` to call the __init__ method of `parent class`. so that we can access the properties of parent class in the child class.

#### Super Keyword

super() keyword is used to access the methods of our parent class from child class.

* Cannot be used outside of class.
* Cannot be used in static method.
* Can only access methods of parent class. Such as `__init__` method, `__str__` method, etc.
* cannot access properties(attributes) of parent class.

```python
# Here we are not following DRY Principle by adding same info again.

class Student(User):
    def __init__(self, student_name, roll_number, age):
        self.student_name = student_name
        self.roll_number = roll_number
        self.age = age

    def __str__(self) -> str:
        return f"Welcome {self.student_name}, Your Roll Number is:{self.roll_number} and Your Age is: {self.age}"
```

```python
s2 = Student("Jane", 36, 20)
print(s2)
```

```python
# now use super() function to call the __init__ method of parent class in the child class.

class Student(User):
    def __init__(self, username: str, id: int, student_name: str, roll_number: int, age: int) -> None:
        super().__init__(username, id)
        self.student_name = student_name
        self.roll_number = roll_number
        self.age = age

    def login(self):
        """"
        Override the login method of parent class to add OTP functionality for student login. and rest all functionality will remain same.
        """
        print("OTP sent for Student Login.")
        super().login()

    def __str__(self) -> str:
        parent_str = super().__str__()
        return f"Welcome {self.student_name}, Your Roll Number is:{self.roll_number} and Your Age is: {self.age}. And your Parent Properties are :{parent_str}"
```

```python
s3 = Student("John", 36, "Jane", 512635, 20)
print(s3)
s3.login()
```
