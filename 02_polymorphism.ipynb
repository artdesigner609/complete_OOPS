## Poly Morphism

Poly Stands for Multiple/Many and Morphism stands for Shape/Form.

So, Polymorphism means many forms. It is the ability of an object to take on many forms. In OOP, it refers to the ability of different classes to be treated as instances of the same class through inheritance. It allows us to use a single interface to represent different types of objects.

For example: Water. it is like if we put in glass or bottle or cup, it will take the shape of the contrainer.

In python we achieve polymorphism by below:

* Method Overloading - where Multiple methods have the same name but different parameters.For example: if we define multiple methods with the same name, the latest definitation replaces all previuos ones.
* Method Overriding
* Operator Overloading

```python
# Let's understand it with the code example. it is also called polymorphism. It + sign has multiple faces to solve the problem, depends upon what we are providing.

print(1+1)  # this + will add them
print("2" + "2")  # this will concetinate them
print([1, 2] + [3, 4])  # this will append to the list
```

```python
# to bypass this in Class, let's see example below:

class User:
    def __init__(self, name: str, mobile_number: int, city: str) -> None:
        if (city == ""):
            self.C1(name, mobile_number)
        else:
            self.C2(name, mobile_number, city)

    def C1(self, name: str, mobile_number: int) -> None:
        self.name: str = name
        self.mobile_number: int = mobile_number

    def C2(self, name: str, mobile_number: int, city: str) -> None:
        self.name = name
        self.mobile_number = mobile_number
        self.city = city
```

```python
user1 = User("raj", 999999, "NY")
print(user1)
```

```python
# improved version from AI:

class User:
    def __init__(self, name: str, mobile_number: int, city: str = "") -> None:
        self.name: str = name
        self.mobile_number: int = mobile_number
        # Only assign city if it is provided
        if city:
            self.city: str = city

    # This controls what happens when you print the object
    def __str__(self) -> str:
        return f"User(name='{self.name}', mobile='{self.mobile_number}', city='{self.city}')"
```

```python
user2 = User("raj", 999999, "NY")
print(user2)
print(user2.name)   # Accessing a specific attribute
```

#### Additional Observations

1. Printing the Object: Your `print(user1)` will currently output something like `<__main__.User object at 0x...内容>`. To see the actual data, add a `__str__ or __repr__` method.
2. Attribute Consistency: In your original `C1` method, `self.city` is never created. If you try to print `user1.city` later for a user without a city, your code will crash with an `AttributeError`. It is safer to set `self.city = None` by default.

#### 2. The Modern Way (Data Classes)

```python
# In Python 3.7+, the @dataclass decorator writes the __init__, __repr__ (printing), and __eq__ (comparing) methods for you automatically.

from dataclasses import dataclass


@dataclass
class User:
    name: str
    mobile_number: int
    city: str = ""  # Default value makes it optional
```

```python
# No need to write __init__ or __str__!
user3 = User("raj", 999999, "NY")
print(user3)
print(user3.name)
print(user3.mobile_number)
print(user3.city)
# Output: User(name='raj', mobile_number=999999, city='NY')
```

### Method Overriding

If a child class has its own method, that is also defined in the parent class, then the method in the child class will override the method in the parent class. This is called method overriding. It allows us to provide a specific implementation of a method that is already defined in its superclass.

```python
class Animal:
    def sound(self) -> str:
        return "Unknow Sound"


class Dog(Animal):
    def sound(self) -> str:
        return "dog is barking"
```

```python
obj = Dog()
# first interpreter initilialize this, and Dog object calles with it method.
print(obj.sound())

obj = Animal()
# then interpreter override the previous obj and initialize Animal class, and called its method.
print(obj.sound())
```
