## Abstraction in Python

Abstraction is derived from the Latin word "abstractus," which means "to draw away" or "to separate." In programming, abstraction is the process of hiding the complex implementation details of a system and exposing only the necessary features to the user. It allows developers to focus on what an object does rather than how it does it.

Another way to understand abstraction is to think of it as a way to create a simplified model of a complex system. By abstracting away the details, we can create a more manageable and understandable representation of the system.

Abstraction -> Abstraction Art -> Hidden message inside Art

For example, we have a mobile phone. We can make a call, sms etc. But in actuality, how this happens we don't know.
Another way to understand is like we have a math library in python. how to calculate sqrt of number, we don't know. its hidden inside the library, we just call it and use it.

Abstract Methods and Abstract Classes

`from abc import ABC, abstractmethod`

we have to ensure 2 things:
1. Inherit from ABC class
2. Should have at least one abstract method

* Abstract classes are useful when we have a group of related objects that should share a common features.
* But should/will have different implementation. 
* Helps in preparing blueprint for other classes.

```python
from abc import ABC, abstractmethod
```

```python
class User(ABC):

    @abstractmethod
    def login(self) -> None:
        pass

    def logout(self) -> None:
        print("logout")

    @abstractmethod
    def auth(self) -> None:
        pass
```

```python
class Buyer(User):
    def login(self) -> None:
        if (self.auth):
            print("User Looged in")
        else:
            print("User Authenticaion failed")

    def check_object(self, link: str) -> None:
        print(link)

    def auth(self):
        return False
```

```python
class Seller(User):
    def login(self) -> None:
        if (self.auth):
            print("Seller Looged in")
        else:
            print("Seller Authenticaion failed")

    def check_object(self, link: str) -> None:
        print(link)

    def auth(self):
        print("OTP")
        return False
```

```python
b1 = Buyer()
b1.login()

s1 = Seller()
s1.login()
```

```python
# Improved version from AI. Hotel cooked a normal taste food in restaurant for all customers.
#
# The Blueprint (Abstract Base Class)
class Food(ABC):
    @abstractmethod
    def taste(self):
        """Every food MUST have a taste defined by the subclass."""
        pass

    def description(self):
        """This is a concrete method. All food shares this logic."""
        print("This is something you can eat. Hotel prepares it for all customers.")

# The Implementation (Concrete Class)


class SpicyFood(Food):
    def taste(self):
        print("This food is very spicy!")


class SweetFood(Food):
    def taste(self):
        print("This food is sweet and sugary.")
```

```python
# Usage
# obj = Food()  # This would now raise an error! (Which is what we want)
my_meal = SpicyFood()  # Prepare the User prefered food(Initialize)
my_meal.description()  # Uses the shared logic
my_meal.taste()        # Uses the specific implementation
```

🥗 Why this is better for learning:
* Enforcement: If you create a new class `ItalianFood(Food)` but forget to add a `taste()` method, Python will stop you immediately. This is the core of abstraction—it guarantees structure.
* Hiding Complexity: The user of your code only needs to know that any Food object has a `.taste()` method. They don't need to care how each specific dish is flavored.
* Consistency: Your original code used `taste()` for the `parent` and `own_taste()` for the `child`. In abstraction, you should use the same method name so that different objects can be used interchangeably (this is also called polymorphism).
