### Constructor & types of Constructor

* Special method which is used for initializing objects with attributes
* Initialize default values of the object, at the time of creation.
* it is __init__ method
* First argument is `self`

There are 3 types of constructors in Python :-

* Parameterized constructor
* Non-Parameterized constructor
* Default Constructor

without constructor:- Object cannot be created.So, in that case default constructor will be called.

#### example of non-parameterized constructor

````python
class Employee:
    def __init__(self) -> None:
        self.age = 21
        self.salary = 10000
````

````python
e1: Employee = Employee()  # example of non-parameterized constructor
e2: Employee = Employee()  # example of non-parameterized constructor

print(e1.__dict__)
print(e2.__dict__)
````

#### example of parameterized constructor

````python
class Employee:
    def __init__(self, age: int, salary: int) -> None:
        self.age: int = age
        self.salary: int = salary
````

````python
e1: Employee = Employee(26, 25000)  # example of parameterized constructor
e2: Employee = Employee(24, 35000)  # example of parameterized constructor

print(e1.__dict__)
print(e2.__dict__)
````

### How to access class members?

* Class members :- Attributes(variables, data) + Actions(Methods, functions)
* We can access these variables using object outside the class.

```python
Syntax :-
Accessing attribute :- object_name.variable_name
Accessing method :- object_name.method_name()
```

[Lecture](https://youtu.be/YDWSCFge0ic?si=Vb2VnX3prIfwawoD)

````python
class Employee:
    def __init__(self, age: int, salary: int) -> None:
        self.age: int = age
        self.salary: int = salary

    def display(self) -> str:
        return f"Your age is: {self.age} and current salary is: {self.salary}"


e1: Employee = Employee(26, 25000)
e2: Employee = Employee(24, 35000)

# it will go into the memory address of il, and get the value of age variable.
print(e1.age)

# it will go into the memory address of e1, and get the value of salary variable.
print(e1.salary)

# we can update the value of variable inside the memory
e1.salary = 75000
print(e1.salary)


# so this e1 memory reference automatically pass in the method() as we have write self for this. there it find the memoery reference of obj(e1 or e2) and execute the method(display) for that object.
print(e1.display())

print(e2.display())  # Your age is: 24 and current salary is: 35000
````

### Builtin Class Functions for OOP

By using these functions, we can `manipulate the data of objects` in a more dynamic way. These functions allow us to check for the existence of attributes, set new values, and delete attributes from objects.

* [Lecture](https://youtu.be/Ua9blALSlKU?si=Kc8TgeWyfuTnhFnN)

Following are built-in class functions :-

* getattr(object_name, attribute_name)
* setattr(object_name, attribute_name, new_value)
* delattr(object_name, attribute_name)
* hasattr(object_name, attribute_name) -> Returns True or False based on whether the object has the specified attribute.

````python
class Employee:
    def __init__(self, name: str, age: int) -> None:
        self.name: str = name
        self.age: int = age


e1 = Employee('raj', 21)
e2 = Employee('jay', 22)

print(getattr(e1, 'age'))  # 21

setattr(e2, 'name', 'jayraj')
print(e2.__dict__)

delattr(e2, 'age')
print(e2.__dict__)

print(hasattr(e1, 'age'))
print(hasattr(e2, 'age'))
````

### Builtin Class attributes

* [Leccture](https://youtu.be/-DwxchTg0VU?si=J-wmS-97WnRjcc8L)

whenever we ceate a python class or object, python automatically creates some attributes for that class or object. These attributes are called built-in class attributes. These attributes provide information about the class or object and can be accessed using the dot notation.

Following are built-in class attributes :-

* __dict__ :- Dictionary containing class's namespace

* __doc__ :- Class documentation string.

* __name__ :- Class Name

* __module__ :- Module name in which class is defined

* __bases__ :- List of base classes -> will be discussed in inheritance lecture

````python
class Employee:
    '''
    This is employee class, for maintaining the Employee data.
    '''

    def __init__(self, name: str, age: int) -> None:
        self.name: str = name
        self.age: int = age


e1 = Employee('raj', 21)
e2 = Employee('jay', 22)

print(Employee.__doc__)
print(Employee.__dict__)

# for example we are writing any condition, where we need the name of that class, so we will use it.
print(Employee.__name__)

# if it will print __main__, it means that class is defined here in current directory. if we import it from other module, then it will print the name of that module
print(Employee.__module__)
````

### `isinstance()` function

for example we have an object and class name, but we don't know whether this object belongs to the class nor not. so, we can use `isinstance()` function to check whether the object belongs to the class or not.

It will return true or Flase.

Generally we use this in the conditional statements. and both arguments are mandatory.

```python
if isinstance(object_name, class_name):
    # do something
```

````python
class Demo:
    pass


d1 = Demo()


class Employee:
    '''
    This is employee class, for maintaining the Employee data.
    '''

    def __init__(self, name: str, age: int) -> None:
        self.name: str = name
        self.age: int = age


e1 = Employee('raj', 21)
e2 = Employee('jay', 22)

print(isinstance(d1, Employee))
print(isinstance(e1, Demo))
````
