## Types of Variables in OOP

### Instance Variables and Instance Methods

In functional programming, we have different types of variables such as local variables, global variables, and nonlocal variables. However, in object-oriented programming (OOP), we have a different classification of variables based on their association with classes and objects. The main types of variables in OOP are:

1. **Instance Variables**: These are variables that are `defined within a class` and are `associated` with an `instance (object)` of that class. 
   1. Each object can have its own unique values for these variables.
   2. They are typically defined within the `__init__` method of a class. Such as name,age and roll number of a Student class.
   3. Instance means object. So, those variables are associated with each object seperately.
   4. modification in one object value, won't affect the other object value.

```python
class Student:

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age
```

````python
class Student:
    def __init__(self, name: str, age: int, rollnumber: int) -> None:
        self.name: str = name
        self.age: int = age
        self.rollnumber: int = rollnumber


student_1 = Student("john", 25, 123)
student_2 = Student("Chelesea", 26, 456)
student_3 = Student("Jane", 27, 789)
````

### Class Variables and Class Methods

1. **Class Variables**: These are variables that are `shared among all instances of a class`. They are `defined within the class` but outside of any instance methods. All instances of the class share the same value for these variables.
   1. It can accessed by reference to the class name. `ClassName.class_variable_name`

```python
class Student:
    # Class variable
    school_name = "ABC High School"

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age

# Accessing class variable
print(Student.school_name)  # Output: ABC High School
```

````python
class Student:
    # Class variable
    school_name = "ABC High School"

    def __init__(self, name: str, age: int) -> None:
        # Instance variables
        self.name: str = name
        self.age: int = age

    def __str__(self) -> str:
        return f"Student {self.name}, aged {self.age} is student of {self.school_name}"


# Accessing class variable
print(Student.school_name)  # Output: ABC High School

# # Updating Class variable
# Student.school_name = 'XYZ High School'
# print("After updating School Name:", Student.school_name)

student_1 = Student("sara", 19)
student_2 = Student("ezabella", 18)

print(student_1)
print(student_2)
````

#### Class Methods

* class methods are methods that are `bound to the class` and `not the instance of the class`. They can modify a class state that applies across all instances of the class.
* First argument is class reference.
* They are defined using the `@classmethod` decorator and take `cls` as the first parameter instead of `self`.
* `self` refers to the instance of the class, while `cls` refers to the class itself. Class methods can be called on the class itself or on an instance of the class.
* as we know that `self` is a variable, as `cls`, so we can name it anything. But for convention, we will use the same variable names.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def get_school_name(cls):
        return cls.school_name
```

````python
class Student:
    # Class variable
    school_name = "ABC High School"

    def __init__(self, name: str, age: int) -> None:
        # Instance variables
        self.name: str = name
        self.age: int = age

    def __str__(self) -> str:
        return f"Student {self.name}, aged {self.age} is student of {self.school_name}"

    @classmethod
    def update_school_name(cls, new_school_name: str) -> str:
        cls.school_name: str = new_school_name
        return cls.school_name


# Accessing class variable old way
print(Student.school_name)  # Output: ABC High School

# Updating Class variable
Student.update_school_name('XYZ High School')
print("After updating School Name:", Student.school_name)

student_1 = Student("sara", 19)
student_2 = Student("ezabella", 18)

print(student_1)
print(student_2)
````

both methohds above shows the same results. then why we use `@classmethod` decorator? why they are important? Why they are necessary?

They produce the `same observable output` in your examples because both approaches change or read the same class attribute, but `@classmethod` exists for specific reasons. Key points:

* What each approach does
  * Direct access (Student.school_name = 'XYZ High School'): immediately sets the class attribute on the class object.
  * `@classmethod update_school_name(cls, ...)`: provides an `encapsulated method` that receives the `class (cls)` and can `modify class-level state` or perform extra logic before/after changing it.
* When classmethod is `useful (advantages)`
  * Encapsulation: centralizes validation, logging, side effects, or computed updates in one place.
  * Subclass friendliness: when called on a subclass, `cls refers to that subclass`, so the method `updates the subclass’s class variable` rather than the `base class’s` — useful if subclasses override or inherit different defaults.
  * Alternative constructors: commonly used to create instances from different data formats (e.g., from_string).
  * Polymorphism: behaves like an instance method but for the class, so subclasses can override it.
  * API clarity: expresses intent that the method operates at the class level.
* When it is unnecessary?
  * Simple scripts where you only set or read a class variable in one place and no validation, logging, or subclass behavior is needed.
  * If you only ever mutate the attribute in one module and do not need the extra indirection.
* Recommendation
  * Use direct attribute access for simple, ad-hoc changes.
  * Use @classmethod when you want encapsulation, validation, subclass-aware behavior, or a clearer public API.

````python
# Short example showing subclass behavior difference:

class Student:
    school_name = "ABC"

    @classmethod
    def update_school_name(cls, name):
        cls.school_name = name


class GradStudent(Student):
    pass


GradStudent.update_school_name("Grad School")
print(Student.school_name)    # still "ABC"
print(GradStudent.school_name)  # "Grad School"
````

`Without classmethod`, assigning `Student.school_name` would affect only `Student` (or whichever name you assign), not automatically target the subclass.

Conclusion: not strictly necessary here, but `classmethods are good practice` when you want `controlled`, subclass-aware class-level operations.

#### `Setters` and `getters` methods

#### Instance Methods

* setter method(mutator method) :- set values of instance variables

* getter method(accessor method) :- get values of instance variables.

````python
class Employee:
    def setName(self, name: str) -> None:  # setter method
        self.name: str = name

    def getName(self) -> str:  # getter method
        print("The name is:", self.name)


e1 = Employee()
e2 = Employee()

e1.setName(input("Enter the name: "))
e2.setName(input("Enter the name: "))

print("e1 object is:", e1 .__dict__)
print("e2 object is:", e1 .__dict__)

e1.getName()
e2.getName()
````

#### Static Method

* Methods which performs operations on external data (data which is not associated with the class or its instances(objects)). See example below.
* it can also perform operations on class data.
* no need to pass object - `self` or Class - `cls` reference as the first parameter.
* created using the `@staticmethod` decorator.

````python
class Bank:
    bank_name = "ECB"  # these are class variables also called static variables
    roi = 3.25  # these are class variables also called static variables

    @staticmethod
    def calculate_simple_interest(principal: int, number_of_years: int) -> str:
        '''
        as we know that roi is fixed for all customers, so it avial it we need to write @staticmethod above the method, so that we could access the value of class(static) variable inside the method.
        '''
        si = (principal * number_of_years * Bank.roi) / 100

        return f"Your due amount is {si}"
````

````python
Bank.calculate_simple_interest(10000, 3)
````
